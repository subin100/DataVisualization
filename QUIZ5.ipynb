I like to see what categories are in my dataset. So lets print first 5 data from our list.

```python
%%bigquery --project datavisualization-338803
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
```

Here, we are going to calculate the second most popular trip.
For this we have to exclude the most popular trip from our dataset and get the highest among remaining data to get second most popular trip.

```python
%%bigquery --project datavisualization-338803
SELECT MAX(start_station_name) as second_most_popular
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name != (SELECT MAX(start_station_name) as most_popular_start_station FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`)
```

So according to the output we got, Ziker Park is the second most popular trip.

Now, lets find out how many trips lasted over an hour and were a round trip.

```python
%%bigquery --project datavisualization-338803
SELECT  COUNTIF(duration_minutes>=60) as over_an_hour_duration ,COUNT( start_station_name) as number_of_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
```

We got the output for the number of trips that last over and hour and number of trips that started and stopped at the same station.

```python
%%bigquery --project datavisualization-338803
SELECT  start_station_name as station_name,  COUNTIF(duration_minutes>=60) as over_an_hour_duration, COUNT( start_station_name) as number_of_trips
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
WHERE start_station_name = end_station_name
GROUP BY start_station_name
```

This is what we were looking for, the trips that are over an hour and trips that started and stopped at the same station and group them by the station name.
