Now, lets import the data from london fire bridage and print the total rows in the dataset.

```python
%%bigquery --project datavisualization-338803
SELECT 
  COUNT(*) as total_rows
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
```

I like to see what categories are in my dataset. So lets print first 10 data from our list.

```python
%%bigquery --project datavisualization-338803
SELECT*
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
limit 10
```

Calculating number of late night false alarm between 10pma and 4am. 

```python
%%bigquery --project datavisualization-338803
SELECT  COUNT(*) as number_of_late_night_false_alarms
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
WHERE incident_group = "False Alarm" AND (hour_of_call >= 22 OR hour_of_call <= 4)
```

Now calculating average number of pumps attending in the london fire brigade.

```python
%%bigquery --project datavisualization-338803
SELECT  incident_group, AVG(num_pumps_attending) as avg_pumps_attending
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY incident_group
```

Finally, printing highest number of pumps of the london fire brigade service calls.

```python
%%bigquery --project datavisualization-338803
SELECT  incident_group, MAX(num_pumps_attending) as Max_pumps_attending
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY incident_group
```
