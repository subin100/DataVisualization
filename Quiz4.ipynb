```python
%%bigquery --project datavisualization-338803
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`

Limit 10
```

This gives us an idea of what is in the table to reference. I have used limit 10 which shows only 10 columns of data and also it takes less amount of time to process. I am more interested on start_station_name and duration_time from this table as we have to work mostly on these two dataset for this assignment.

```python
%%bigquery --project datavisualization-338803
SELECT AVG(table.duration_minutes) as average_duration_minutes, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` as table
GROUP BY start_station_name
```

Here, we calculated average time for the trip using Avg command and grouped it in a average_duration_minutes and then we based in on starting point. For example, according to the table, it takes around 27.17 minutes to make a trip in Ziker park West. 

```python
%%bigquery --project datavisualization-338803
WITH t as(
SELECT AVG(table.duration_minutes) as average_duration_minutes, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` as table
GROUP BY start_station_name
)
SELECT *, 
    RANK() OVER(
      PARTITION BY start_station_name
    ORDER BY start_station_name) rn_num
FROM t
ORDER BY rn_num
```

From the table it seems that lots of the trips at the each starting point only once.
