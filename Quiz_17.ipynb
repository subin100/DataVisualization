```python
import pandas as pd
import statistics

iris_df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/iris.csv')
iris_df
```

##Does converting SepalLength to integer increase or decrease the mean?

```python
from numpy.ma.core import mean
mean(iris_df.SepalWidth)
```

Now, converting to integer.

```python
new_df= iris_df.SepalWidth.astype('int') 
new_df
```

```python
mean(new_df)
```

Hence, it decreases the mean.

##Does the direction of the shift remain the same if you groupby Class?
