```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df
```

```python
ax = sns.scatterplot(data = df, 
                x = 'workout_type', 
                y = 'distance',
                size = 'calories',
                hue = 'workout_type',
                legend = 'brief')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1.1,1))
```

```python
sns.set(rc={"figure.figsize":(20, 10)})
sns.heatmap(df.corr(), annot=True, linewidths=1.5)
```

```python
sns.set(rc={"figure.figsize":(10, 10)})
ax = sns.scatterplot(data = df, 
                x = 'distance', 
                y = 'time',             
                size = 'distance',
                hue = 'workout_type',
                x_bins = "avg_speed")
sns.move_legend(ax, "upper right", bbox_to_anchor=(1.25,1))
```
